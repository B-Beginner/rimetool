```python
from bs4 import BeautifulSoup
import ebooklib
from ebooklib import epub
import re

# Load the EPUB file
book = epub.read_epub(r'J:\Project_Playground\rime2\rimetool\examples\jinkuiyaolue.epub')

# # Iterate through the items in the book
# for item in book.get_items():
#     if item.get_type() == ebooklib.ITEM_DOCUMENT:
#         print(item.get_body_content().decode('utf-8'))



# Open a file to write the content
# with open(r'J:\Project_Playground\rime2\rimetool\examples\jkyl2.txt', 'w', encoding='utf-8') as f:
#     # Iterate through the items in the book
#     for item in book.get_items():
#         if item.get_type() == ebooklib.ITEM_DOCUMENT:
#             content = item.get_body_content().decode('utf-8')
#             # Find all sections with id="CHP"
#             sections = re.findall(r'<.*?id="CHP.*?>.*?</.*?>', content, re.DOTALL)
#             print(sections)
#             print("\n\n")
#             for section in sections:
#                 f.write(section)
#                 f.write('\n')  # Add a newline for separation between sections

with open(r'J:\Project_Playground\rime2\rimetool\examples\jkyl2.txt', 'w', encoding='utf-8') as f:
    # 遍历书籍中的所有项目
    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            # 获取文档内容并解码为 UTF-8 字符串
            content = item.get_body_content().decode('utf-8')
            # 使用 BeautifulSoup 解析 HTML 内容
            soup = BeautifulSoup(content, 'html.parser')
            # 查找所有 id 以 CHP 开头的 div 标签
            sections = soup.find_all('div', id=lambda x: x and x.startswith('CHP'))
            print(sections)
            for section in sections:
                # 找到当前 section 标签及其后续的所有兄弟标签
                elements = [section] + list(section.next_siblings)
                for element in elements:
                    if isinstance(element, type(soup.new_tag('div'))):
                        # 如果是 div 标签，则写入其 HTML 字符串
                        f.write(str(element))
                        f.write('\n')  # 添加换行符以分隔不同的部分
                    else:
                        # 如果遇到另一个 id 以 CHP 开头的 div 标签，则停止
                        if isinstance(element, type(soup.new_tag('div'))) and element.get('id', '').startswith('CHP'):
                            break
```

```python
from bs4 import BeautifulSoup
with open(r'J:\Project_Playground\rime2\rimetool\examples\jkyl2.txt', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content, 'html.parser')

# 打开一个文件以写入提取的内容
with open(r'J:\Project_Playground\rime2\rimetool\examples\jkyl3.txt', 'w', encoding='utf-8') as output_file:
    # # 删除所有img标签
    # for img_tag in soup.find_all('img'):
    #     img_tag.decompose()

    # 查找所有符合条件的h3标签

    h3_tags = soup.find_all('div', class_='h3', id=re.compile(r'CHP\d+'))
    for tag in h3_tags:
        output_file.write(str(tag))
        output_file.write('\n')

    # 查找所有符合条件的p标签
    p_tags = soup.find_all('div', class_='pCls')
    for tag in p_tags:
        output_file.write(str(tag))
        output_file.write('\n')
```
