# Import Required Libraries
Import the necessary libraries, such as os and pandas.

```python
# Import the necessary libraries
import os
import pandas as pd
import re
```

# Read the File
Use Python's built-in functions to read the content of the file 'jinkuiyaolue_extract_without_div_v2.txt'.

```python
# Read the File
file_path = r'J:\Project_Playground\rime2\rimetool\rimetool\epub\金匮要略 中医临床必读丛书4.txt'

# Check if the file exists
if os.path.exists(file_path):
    # Open and read the file
    with open(file_path, 'r', encoding='utf-8') as file:
        file_content = file.read()
else:
    file_content = None

# Display the content of the file
file_content
```

# Display File Content
Print the content of the file to verify it has been read correctly.

```python
# Display File Content
if file_content:
    print(file_content)
else:
    print("File not found or is empty.")
```

# 拆分成短句


```python

# Split content by punctuation, spaces, and newlines
# 拆分短句的方法：先用标点符号拆
split_content = re.split(r'[，。！？；：、\s\n]+', file_content)

# Write each split content on a new line in the output file
with open(r'J:\Project_Playground\rime2\rimetool\rimetool\epub\金匮要略 中医临床必读丛书5拆分.txt', 'w', encoding='utf-8') as output_file:
    for item in split_content:
        if item:  # Avoid writing empty strings
            output_file.write(item + '\n')
```

# 处理长句

```python
# 拆分长句的方法：单纯地把所有标点符号去除
with open(r'J:\Project_Playground\rime2\rimetool\rimetool\epub\金匮要略 中医临床必读丛书6拆分长.txt', 'w', encoding='utf-8') as output_file:
    for item in file_content.split('\n'):
        if item:  # Avoid writing empty strings
            cleaned_item = re.sub(r'[^\w]', '', item)  # Remove punctuation and spaces
            output_file.write(cleaned_item + '\n')
```
